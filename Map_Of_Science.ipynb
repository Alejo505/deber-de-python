```python
import re
import csv
import pandas as pd
from io import StringIO
from pandas import DataFrame
import matplotlib.pyplot as plt
```

```python
with open("pubmed_scherichia.txt", "r", encoding="utf8") as f:
    my_text = f.read()
print(len(my_text))
my_text = re.sub(r'\n\s{6}', ' ', my_text)
```

```python
#Minado de datos usando expresiones regulares para las direcciones por país.
country1 = re.findall(r'A\D.*.\s\s\d{5}, USA', my_text)
país1=[]
for i in country1:
    i = i.replace("USA","US")
    país1.append(i)
else:
    print(país1[:10])
```

```python
country1 = re.findall(r'A\D.*.\s\s\d{5}, (USA)', my_text)
país1=[]
for i in country1:
    i = i.replace("USA","US")
    país1.append(i)
else:
    print(país1[:10])
    
    print(len(país1))
    
país2 = re.findall(r'A\D.*., (UK) ', my_text)
print(país2[:20])

print(len(país2))

country3 = re.findall(r'A\D.*., (Brazil)', my_text)
país3=[]
for i in country3:
    i = i.replace("Brazil","BR")
    país3.append(i)
else:
    print(país3[:10])
    
    print(len(país3))
    
country4 = re.findall(r'A\D.*., (China)', my_text)
país4=[]
for i in country4:
    i = i.replace("China","CN")
    país4.append(i)
else:
    print(país4[:10])
    
    print(len(país4))

country5 = re.findall(r'A\D.*., (Australia)', my_text)
país5=[]
for i in country5:
    i = i.replace("Australia","AU")
    país5.append(i)
else:
    print(país5[:10])
    
    print(len(país5))
    
country6 = re.findall(r'A\D.*., (Norway)', my_text)
país6=[]
for i in country6:
    i = i.replace("Norway","NO")
    país6.append(i)
else:
    print(país6[:10])
    
    print(len(país6))
```
