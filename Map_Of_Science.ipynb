```python
import re
import csv
import pandas as pd
from io import StringIO
from pandas import DataFrame
import matplotlib.pyplot as plt
```

```python
with open("pubmed_scherichia.txt", "r", encoding="utf8") as f:
    my_text = f.read()
print(len(my_text))
my_text = re.sub(r'\n\s{6}', ' ', my_text)
```

```python
codes1 = re.findall(r'A\D.*.\s\s\d{5}, USA', my_text)
zipcodes1=[]
for i in codes1:
    i = i.replace("USA","US")
    zipcodes1.append(i)
else:
    print(zipcodes1[:10])
```

```python
codes1 = re.findall(r'A\D.*.\s\s\d{5}, (USA)', my_text)
zipcodes1=[]
for i in codes1:
    i = i.replace("USA","US")
    zipcodes1.append(i)
else:
    print(zipcodes1[:10])
    
    print(len(zipcodes1))
    
zipcodes2 = re.findall(r'A\D.*., (UK) ', my_text)
print(zipcodes2[:20])

print(len(zipcodes2))

codes3 = re.findall(r'A\D.*., (Brazil)', my_text)
zipcodes3=[]
for i in codes3:
    i = i.replace("Brazil","BR")
    zipcodes3.append(i)
else:
    print(zipcodes3[:10])
    
    print(len(zipcodes3))
    
codes4 = re.findall(r'A\D.*., (China)', my_text)
zipcodes4=[]
for i in codes4:
    i = i.replace("China","CN")
    zipcodes4.append(i)
else:
    print(zipcodes4[:10])
    
    print(len(zipcodes4))

codes5 = re.findall(r'A\D.*., (Australia)', my_text)
zipcodes5=[]
for i in codes5:
    i = i.replace("Australia","AU")
    zipcodes5.append(i)
else:
    print(zipcodes5[:10])
    
    print(len(zipcodes5))
    
codes6 = re.findall(r'A\D.*., (Norway)', my_text)
zipcodes6=[]
for i in codes6:
    i = i.replace("Norway","NO")
    zipcodes6.append(i)
else:
    print(zipcodes6[:10])
    
    print(len(zipcodes6))
```

```python
zip_count1 = []
for z in zipcodes1:
     zip_count1.append(zipcodes1.count(z))

len(zip_count1)
```
